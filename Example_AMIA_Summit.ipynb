```python
# Install the transformers library
!pip install transformers

# Import necessary libraries
import tensorflow as tf
from transformers import TFGPT2LMHeadModel, GPT2Tokenizer

# Load the GPT-2 model and tokenizer
model_name = "gpt2"  # You can use "gpt2-medium", "gpt2-large", etc. for larger models
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = TFGPT2LMHeadModel.from_pretrained(model_name, pad_token_id=tokenizer.eos_token_id)

# Function to generate text
def generate_text(prompt, max_length=100, temperature=0.7, top_k=40):
    # Encode the input prompt
    input_ids = tokenizer.encode(prompt, return_tensors="tf")

    # Generate text
    output = model.generate(
        input_ids,
        max_length=max_length,
        temperature=temperature,
        top_k=top_k,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )

    # Decode the generated text
    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
    return generated_text

# Print a separator line
print("\n\n")
print("=" * 50)

# Generate a response to the given statement
prompt = "I am the primary care provider. Because the patient has reported the following list of symptoms. Chills Cough Shortness of breath, Fatigue, Muscle and body aches. Headache, New loss of taste, Sore throat, Congestion. I have diagnosed them with the disease: "
response = generate_text(prompt)
print(response)

# Print a separator line
print("=" * 50)
```

```python
# Install the transformers library
!pip install transformers

# Import necessary libraries
import tensorflow as tf
from transformers import TFGPT2LMHeadModel, GPT2Tokenizer
import numpy as np

# Load the GPT-2 model and tokenizer
model_name = "gpt2"  # You can use "gpt2-medium", "gpt2-large", etc. for larger models
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = TFGPT2LMHeadModel.from_pretrained(model_name, pad_token_id=tokenizer.eos_token_id)

# Function to get next token probabilities
def get_next_token_probs(prompt, top_k=10):
    # Encode the input prompt
    input_ids = tokenizer.encode(prompt, return_tensors="tf")

    # Get the model's logits (raw output scores)
    outputs = model(input_ids)
    logits = outputs.logits[:, -1, :]  # Get logits for the last token

    # Apply softmax to convert logits to probabilities
    probs = tf.nn.softmax(logits, axis=-1).numpy().flatten()

    # Get the top-k tokens and their probabilities
    top_k_indices = np.argsort(probs)[-top_k:][::-1]  # Indices of top-k tokens
    top_k_tokens = [tokenizer.decode([idx]) for idx in top_k_indices]  # Decode tokens
    top_k_probs = probs[top_k_indices]  # Probabilities of top-k tokens

    return list(zip(top_k_tokens, top_k_probs))

# Print a separator line
print("\n\n")
print("=" * 50)

# Get and print the next token probabilities
prompt = "I am the primary care provider. Because the patient has reported the following list of symptoms. Chills Cough Shortness of breath, Fatigue, Muscle and body aches. Headache, New loss of taste, Sore throat, Congestion. I have diagnosed them with"
next_token_probs = get_next_token_probs(prompt, top_k=10)

# Print bold headers
print("\033[1mTop-K Next Tokens and Their Probabilities:\033[0m")
for token, prob in next_token_probs:
    print(f"\033[1mToken:\033[0m {token:<20} \033[1mProbability:\033[0m {prob:.4f}")

# Print a separator line
print("=" * 50)
```
